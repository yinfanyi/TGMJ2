```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_linspace(x, interval, num):
    lower_bound = x - interval
    if lower_bound < 0:
        lower_bound = 0
    upper_bound = x + interval
    values = np.linspace(lower_bound, upper_bound, num)
    return values
# 选择的值  
x = 60000  

# 设置范围  
lower_bound = x - 10000  # 50000  
upper_bound = x + 10000  # 70000  

# 生成 50 个均匀间隔的数值  
values = np.linspace(lower_bound, upper_bound, 50)  

# 打印结果  
print(values)
print(get_linspace(x, 10000, 50))
```

```python
import os
import re

output_dir = './k1'
output_files = os.listdir(output_dir)
n = 100
f1,f21,z,k1,k2,k3,k4,k5,k6,k7=[1.08570230e+02, 2.15525292e+02, 8.81208515e-01, 6.54814411e+05,
                        6.00253605e+05, 4.49265232e+04, 2.31510442e+05, 2.34302564e+05,
                        8.01972959e+05, 3.05995422e+02]
# list = get_linspace(k1, 10000, 100)
list = get_linspace(512930, 2, 50)
list = np.linspace(100, 1000000, 100)  #k1

highest_data = []
file_name = []
for file in output_files:
    match = re.search(r'(\d+).csv', file)
    if match:  
        # 提取数字并转换为整数  
        i = int(match.group(1))
        data = pd.read_csv(os.path.join(output_dir, file))
        highest_point = data.loc[data['outside_z_pos_datas'].idxmax()]
        value = list[i] 
        highest_point['value'] = value
        highest_data.append(highest_point)
        file_name.append(file)
highest_data_df_mujoco = pd.DataFrame(highest_data)
highest_data_df_mujoco['file_name'] = file_name
highest_data_df_mujoco.to_csv(f'{output_dir}/highest_data.csv', index=False)
plt.scatter(highest_data_df_mujoco['value'], highest_data_df_mujoco['outside_z_pos_datas'], label='mujoco',s=2)
plt.grid()
plt.legend()
plt.xlabel('k1')
plt.ylabel('height')
plt.show()
```
